```python
def process_timestamp(df):
    for index, row in df.iterrows():
        try:
            # Convert 'Timestamp' column to datetime format
            timestamp = pd.to_datetime(row['Timestamp'], format='%d/%m/%Y', errors='coerce')

            # Check if conversion resulted in NaT (Not a Time)
            if pd.isna(timestamp):
                raise ValueError(f"Invalid timestamp format: {row['Timestamp']}")

            # Update 'Timestamp' in the DataFrame
            df.at[index, 'Timestamp'] = timestamp
            
        except Exception as e:
            print(f"Error processing timestamp at index {index}: {e}")
            # Drop row if there was an error processing
            df.drop(index, inplace=True)
    
    # Sort DataFrame by Timestamp (ignoring NaT)
    df.sort_values(by='Timestamp', inplace=True, na_position='last')
    
    # Remove NaT rows
    df.dropna(subset=['Timestamp'], inplace=True)
    return df
```
